```
import cv2 as cv

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import tensorflow

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

plt.figure(figsize=(11,6), facecolor='w')

plt.subplot(241)
img = cv.imread('images/12/Lena.png', 0)
plt.imshow(img, cmap='gray')
plt.title(u'原始图片')

plt.subplot(242)
image = img[100:190, 100:180]
plt.imshow(image, cmap='gray')
plt.title(u'脸部图片')

plt.subplot(243)
res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
plt.imshow(res, cmap='gray')
plt.title(u'放大图片')

plt.subplot(244)
height, width = img.shape[:2]
res = cv.resize(img,(int(0.5*width), int(0.5*height)), interpolation = cv.INTER_CUBIC)
plt.imshow(res, cmap='gray')
plt.title(u'缩小图片')

plt.subplot(245)
H = np.float32([[1,0,100],[0,1,50]])
rows,cols = img.shape[:2]
res = cv.warpAffine(img,H,(rows, cols)) #需要图像、变换矩阵、变换后的大小
plt.imshow(res, cmap='gray')
plt.title(u'平移')

plt.subplot(246)
M = cv.getRotationMatrix2D((cols/2,rows/2),45,1)
#第三个参数：变换后的图像大小
res = cv.warpAffine(img,M,(rows,cols))
plt.imshow(res, cmap='gray')
plt.title(u'旋转')

# 仿射就是旋转加拉伸
plt.subplot(247)
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
#第三个参数：变换后的图像大小
res = cv.warpAffine(img,M,(rows,cols))
plt.imshow(res, cmap='gray')
plt.title(u'仿射')

plt.subplot(248)
pts1 = np.float32([[56,65],[238,52],[28,237],[239,240]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
M = cv.getPerspectiveTransform(pts1,pts2) #透射就是将图片投影到新的平面上
res = cv.warpPerspective(img,M,(200,200))
plt.imshow(res, cmap='gray')
plt.title(u'透射')

plt.show()
```

```
#线段操作
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = np.zeros((512,512,3),np.uint8)#生成一个空彩色图像
cv2.line(img,(0,0),(511,511),(55,255,155),5)
plt.imshow(img,'brg')
plt.show()
```

```
#绘制圆形
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = np.zeros((512,512,3),np.uint8)#生成一个空彩色图像
cv2.circle(img,(200,200),50,(55,255,155),5)#修改最后一个参数
plt.imshow(img,'brg')
plt.show()
```

```
#绘制矩形
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = np.zeros((512,512,3),np.uint8)#生成一个空彩色图像
cv2.rectangle(img,(20,20),(411,411),(55,255,155),5)
plt.imshow(img,'brg')
plt.show()
```

```
#绘制椭圆
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = np.zeros((512,512,3),np.uint8)#生成一个空彩色图像
cv2.ellipse(img,(256,256),(150,100),0,0,360,250,5)
#注意最后一个参数-1，表示对图像进行填充，默认是不填充的，如果去掉，只有椭圆轮廓了
plt.imshow(img,'brg')
plt.show()
```
