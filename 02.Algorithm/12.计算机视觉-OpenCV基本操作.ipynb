```
import cv2 as cv

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import tensorflow

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 
mpl.rcParams['axes.unicode_minus'] = False # '-'

plt.figure(figsize=(11,6), facecolor='w')

plt.subplot(241)
img = cv.imread('images/12/Lena.png', 0)
plt.imshow(img, cmap='gray')
plt.title(u'')

plt.subplot(242)
image = img[100:190, 100:180]
plt.imshow(image, cmap='gray')
plt.title(u'')

plt.subplot(243)
res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
plt.imshow(res, cmap='gray')
plt.title(u'')

plt.subplot(244)
height, width = img.shape[:2]
res = cv.resize(img,(int(0.5*width), int(0.5*height)), interpolation = cv.INTER_CUBIC)
plt.imshow(res, cmap='gray')
plt.title(u'')

plt.subplot(245)
H = np.float32([[1,0,100],[0,1,50]])
rows,cols = img.shape[:2]
res = cv.warpAffine(img,H,(rows, cols)) #
plt.imshow(res, cmap='gray')
plt.title(u'')

plt.subplot(246)
M = cv.getRotationMatrix2D((cols/2,rows/2),45,1)
#
res = cv.warpAffine(img,M,(rows,cols))
plt.imshow(res, cmap='gray')
plt.title(u'')

# 
plt.subplot(247)
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
#
res = cv.warpAffine(img,M,(rows,cols))
plt.imshow(res, cmap='gray')
plt.title(u'')

plt.subplot(248)
pts1 = np.float32([[56,65],[238,52],[28,237],[239,240]])
pts2 = np.float32([[0,0],[200,0],[0,200],[200,200]])
M = cv.getPerspectiveTransform(pts1,pts2) #
res = cv.warpPerspective(img,M,(200,200))
plt.imshow(res, cmap='gray')
plt.title(u'')

plt.show()
```
