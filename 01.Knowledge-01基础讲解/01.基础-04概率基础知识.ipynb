#概率基础


#1 先验概率
P(A)：事件${A_1,A_2,...,A_n}$看作是导致时间B发生的原因，在不知道事件B是否发生的情况下，它们的概率是${P(A_1),P(A_2),...,P(A_n)}$，通常称为先验概率


#2 后验概率-条件概率
P(A|B)：如果B已经发生，那么对$A_1,A_2,...A_n$发生的可能性大小$P(A_1|B),P(A_2|B),...,P(A_n|B)$有了新的估价
$$P(A|B) = \frac{P(AB)}{P(B)}$$


#3 乘法公式
- 如果A、B是两个独立的事件(A的发生不会影响B的发生),那么就有$P(AB) = P(A) * P(B)$
- 如果A、B是两个不独立的事件(A的发生会影响B的发生),那么就有$P(AB)=P(B|A)P(A)=P(A|B)P(B)$
比如抽奖，一共有100个人，只有10个奖，那么抽完放回抽奖箱和抽完不放回抽奖箱，是完全不一样的，前者会导致独立事件的发生，后者则会导致非独立事件的发生，设
A：第一个人抽中奖，B：第二个人抽中奖，C：第三个人抽中奖
解：那么就有$P(A) = \frac{10}{100} = 0.1$。如果放回抽奖箱，那么$P(B) = \frac{10}{100} = 0.1$,如果不放回抽奖箱，那么P(B)的计算就不简单了，P(B)应该是$P(B|A) + P(B|!A) = \frac{19}{99}$,那么P(AB)是多少呢，如果独立，那么A、B同时中奖的概率就是$P(A) * P(B) = 0.1 * 0.1 = 0.01$，如果不独立呢，$P(AB) = P(A) * P(B|A) = 0.1 * \frac{1}{11}$

乘法公式的推广：
- 独立事件：$P(A_1)P(A_2)...P(A_n)$
- 非独立事件$P(A_1A_2...A_n)=P(A_1)P(A_2|A_1)P(A_3|A_1A_2)...P(A_n|A_1A_2...A_{n-1})$


#4 加法公式
$$P({A}\cup{B})=P(A)+P(B)-P(A\cap{B})$$


#5 全概率公式
$$P(B)=P(A_1B+A_2B+...+A_nB)=\sum_{i=1}^nP(A_iB)=\sum_{i=1}^nP(A_i)P(B|A_i)$$
B和A不是一个级别的东西
设A为市场占有率,B为次品率，本例题求解的就是P(B),根据题目得到以下概率：

$P(A_1) = 0.3$,$P(A_2) = 0.2,P(A_3) = 0.5$

第一家工厂的次品率:$P(B|A_1) = 0.03$

第二家工厂的次品率:$P(B|A_2) = 0.03$

第三家工厂的次品率:$P(B|A_3) = 0.01$

那么$P(B) = P(A_1)P(B|A_1)+P(A_2)P(B|A_2)+P(A_3)P(B|A_3) = 0.3 * 0.03 + 0.2 * 0.03 + 0.5 * 0.01 = 0.09 + 0.06 + 0.05 = 0.2$

##7.1 期望


###7.1.1 概念描述
在概率论和统计学中，数学期望(mean)（或均值，亦简称期望）是试验中每次可能结果的概率乘以其结果的总和。是最基本的数学特征之一。它反映随机变量平均取值的大小。$E(X)=E(x_1)+E(x_2)+...+E(x_n)=x_1p_2+x_2p_2+...+x_np_n=\sum_{i=1}^nx_ip_i=\sum_{i=1}^nx_nf(x_n)$其实就是n*p

无条件成立：E(kX)=kE(X), E(X+Y)=E(X)+E(Y)

如果A，B独立，则P(A|B)=P(A),E(AB)=E(A)E(B)


###7.1.2 举例
异或问题

从1，2，3，...,98，99，2015这100个数中任意选择若干个数(可能为0个数)求异或，试求其期望

解：针对任何一个二进制位，取奇数个1异或后会得到1，取偶数个1异或后会得到0，与取0的个数无关。给定的最大数为2015=11111011111，共11位。针对每一位分别计算，考虑第i位$X_i$，假定给定的100个数中第i位一种有n个1，m个0，某次采样取到的1的个数为k，有：

$P(X_i=1)=\frac{2^m \bullet \sum_{k \in odd}C_n^k}{2^{m+n}}=\frac{\sum_{k \in odd}C_n^k}{2^n}=\frac{1}{2}$，这就说明11位二进制数中，每个位取1的期望都是0.5，根据$E(X+Y)=E(X)+E(Y)$，则有

$E(X)=E[\sum_{i=0}^{10}(x_i \bullet P(X_i))]=E[\sum_{i=0}^{10}(2^i \bullet P(X_i=1) + 0 \bullet P(X_i=0))]=E[\sum_{i=0}^{10}(2^iP(X_i=1))]$

$=\sum_{i=0}^{10}E(2^iP(X_i=1))=\sum_{i=0}^{10}2^i \bullet E(P(X_i=1))=\sum_{i=0}^{10}2^i \bullet \frac{1}{2}=\frac{1}{2}\sum_{i=0}^{10}2^i=\frac{2^0+2^1+...+2^{10}}{2}=1023.5$

如果将2015改成1024，那么总期望就是575.5
- 如果“n个1”退化成“没有1”，即候选的数都是0，则取到1为奇数是不可能的，概率为0。（这种情况在2015这个例子中不存在，但在思考题1024中是存在的。）
- 如果“n个1”中的n不为0，那么，题目就是最正常的计算“给定n个1、m个0，从这(n+m)个数中取到奇数个1的概率”，m是否为0不影响分析。
此时，先取到k个1，再随便取若干个0。
1. 从n个1中取k个1的取法，一共有$C_n^k$种取法；
2. 从n个1中取数的所有取法为$2^n$；
因此，取到k个的概率为$\frac{C_n^k}{2^n}$；
k可以有从0到n的n+1种候选，但我们只关心k取奇数的情况，从而：
k取奇数的概率为：$\sum_k\frac{C_n^k}{2^n}$,k为奇数，根据二项式的性质：所有奇数项和偶数项和相等，得到该式子为$\frac{1}{2}$。
而显然，m是几不关心，上面的$\frac{1}{2}$就是最终结论。

Hash问题

某Hash函数将任一字符串非均匀映射到正整数k，概率为$2^{-k}$，$P\{Hash(<string>)=k\}=2^{-k}$。现有字符串集合S，其元素经映射后，得到的最大整数为10.试估计S的元素个数

由于Hash映射成整数是指数级衰减的，"最大整数为10"这一条件可近似考虑成"整数10曾经出现",继续近似成"整数10出现过一次"。字符串被映射成10的概率为$p=2^{-10}=\frac{1}{1024}$，从而一次映射即两点分布$\begin{cases}
P(X=1)=\frac{1}{1024}\\
P(X=0)=\frac{1023}{1024}
\end{cases}$，从而n个字符串的映射，即二项分布：$P{X=k}=C_n^kp^k(1-p)^{n-k}$，其中$p=\frac{1}{1024}$.
二项分布的期望是$E(\{P=K\})=np$，其中$p=\frac{1}{1024}$，而期望表示n次时间发生的次数，当前问题中发生了一次，从而$np=1 \Rightarrow n=\frac{1}{p} \Rightarrow n=1024$

###7.1.3 随机变量的函数的期望
定理：设随机变量Y是随机变量X的函数$Y=g(X)$，这里g是连续函数，那么

1. 若X是离散型随机变量，且X的概率分布为$P\{X=x_i\}=p_i, i=1,2,...,n$,则Y的期望$E(Y)=E[g(X)]=\sum_{i=1}^ng(x_i)p_i$

2. 若X是连续型随机变量，且概率密度为f(x),则Y的期望$E(Y)=E[g(x)]=\lmoustache_{-\infty}^{+\infty}g(x)f(x)dx$

##7.2 方差
$D(X)=E[(X-E(X))^2]=E(X^2)-[E(X)]^2$


D(c)=0，常数项求期望是0

D(X+c)=D(X)

$D(kX)=k^2D(X)$

X和Y独立，则有D(X+Y)=D(X)+D(Y)


##7.3 协方差
Cov(X,Y)=E{[X-E(X)][Y-E(Y)]}

Cov(X,Y)=Cov(Y,X)

Cov(aX+b, cY+d)=acCov(X,Y)

D(X) = Cov(X,X)

$Cov(X_1+X_2,Y)=Cov(X_1,Y)+Cov(X_2,Y)$

$Cov(X,Y)=E(XY)-E(X)E(Y)$

如果X，Y独立，则Cov(X,Y)=0。如果协方差为零，我们定义为不相关

###7.3.1 协方差的意义
协方差是两个随机变量具有想同方向变化趋势的度量
- 若$Cov(X,Y)>0$，它们的变化趋势相同
- 若$Cov(X,Y)<0$，它们的变化趋势相反
- 若$Cov(X,Y)=0$，它们不相关，那么就说明它们线性不相关(一阶不相关)

###7.3.2 协方差上界定理
若$D(X)=\sigma_1^2$，$D(Y)=\sigma_2^2$，则$|Cov(X,Y)| \leq \sigma_1\sigma_2$，当且仅当X和Y之间有线性关系时，等号成立

证明：取任意实数t，构造随机变量Z，$Z=(X-E(X)) \bullet t + (Y-E(Y))$，从而$E(Z^2)=\sigma_1^2t^2+2Cov(X,Y) \bullet t+\sigma_2^2$

$\because Z^2 \geq 0, \therefore E(Z^2) \geq 0$，即$\sigma_1^2t^2+2Cov(X,Y) \bullet t+\sigma_2^2 \geq 0$，则它的判别式$\Delta=4Cov^2(X,Y)-2\sigma_1^2\sigma_2^2 \leq 0$

$\therefore |Cov(X,Y)| \leq \sigma_1\sigma_2$

当X，Y成线性关系的时候，即$X=\alpha{Y}+b$, 则$|Cov(X,Y)|=\sigma_1\sigma_2$

##7.4 Pearson相关系数

###7.4.1 定义
$\rho_{XY}=\frac{Cov(X,Y)}{\sqrt{D(X)D(Y)}}$，由协方差上界定理可知, $|\rho| \leq 1$，当且仅当X与Y有线性关系时，等号成立

相关系数是标准尺度下的协方差。上面关于协方差与XY相互关系的结论，完全适用于相关系数和XY的相互关系。若$\rho_{XY}=0$，则X，Y是不相关的

##7.5 协方差矩阵
对于n个随机变量$(X_1,X_2,...,X_n)$，任意两个元素$X_i$和$X_j$都可以得到一个协方差，从而形成$n*n$的矩阵

$$
C=\begin{bmatrix}
c_{11}&c_{12}&\dots&c_{1n}\\
c_{21}&c_{22}&\dots&c_{2n}\\
\vdots&\vdots&\ddots&\vdots\\
c_{n1}&c_{n2}&\dots&c_{nn}
\end{bmatrix}
$$

由于$c_{ij}=E{[X_i-E(X_i)][X_j-E(X_j)]}=Cov(X_i,X_j)$，所以协方差矩阵是一个对称矩阵，而且是一个正定矩阵

假设有如下数列 1，1，3，4，6，8，9，199
- 均数：也叫平均数，算数平均数，就是所有数字的和除以数字的个数，主要考察数列的集中趋势。均值为$\frac{231}{7} = 33$
- 中位数：数列按照从小到大排序，然后获取中间的那个数字就是中位数，如果中间的是两个数字，那么中位数就是这两个数的平均数，主要也是考察数列的集中趋势,中位数为6
- 众数：数列中出现最多次数的数字，众数为1
- 极差：数列中最大的数减去最小的数。考察数列中的数的紧凑程度，极差为198
- 中程数：数列中最大数和最小数的平均值，主要考察数列的集中趋势，中程数为99
- 总体均值$\mu=\frac{\sum_{i=1}^Nx_{i}}{N}$
- 总体方差$\sigma^2=\frac{\sum_{i=1}^N(x_{i}-\mu)^2}{N}$
总体方差也可以用这个公式计算$\sigma^2=\frac{\sum_{i=1}^Nx_{i}^2}{N}-\mu^2$
所以上述队列的总体方差就是$\frac{2048+900+841+729+625+576+38416}{8}=5516.875$
- 样本均值$\overline{x}=\frac{\sum_{i=1}^nx_{i}}{n}$
- 样本方差$S^2=\frac{\sum_{i=1}^n(x_{i}-\overline{x})^2}{n-1}$
样本方差使用n-1是为了保证无偏
- 标准差$\sigma=\sqrt{\sigma^2}$

描述数列集中趋势可以使用均数、中位数、众数和中程数，但是这不能分好的反应数列中的数的发散程度。

##7.5 相对熵
又称互熵，交叉熵，鉴别信息，Kullback熵，Kullback-Leible散度

设$p(x),q(x)$是X中取值的两个概率分布，则p对q的相对熵是$D(p||q)=\sum_xp(x)log\frac{p(x)}{q(x)}=E_{p(x)}log\frac{p(x)}{q(x)}$

- 相对熵可以度量两个随机变量的距离
- 一般的,$D(p||q) \neq D(q||p)$
- $D(p||q) \geq 0,D(q||p) \geq 0$

##7.6 互信息
两个随机变量X，Y的互信息，定义为X，Y 的联合分布和独立分布乘积的相对熵

$I(X,Y)=D(P(X,Y) || P(X)P(Y))=\sum_{x,y}p(x,y)log\frac{p(x,y)}{p(x)p(y)}$

##7.7 信息增益
信息增益表示得知特征A的信息而使得类X 的信息的不确定性减少的程度.

定义:特征A对训练数据集D的信息增益g(D,A)，定义为集合D的经验熵H(D)与特征A给定条件下D的经验条件熵H(D|A)之差。g(D,A)=H(D) – H(D|A).显然，这即为训练数据集D和特征A的互信息
